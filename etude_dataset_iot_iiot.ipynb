```python
#fonction d'encodage des variable
def encode_text_dummy(df, name):

    dummies = pd.get_dummies(df[name])

    for x in dummies.columns:

        dummy_name = f"{name}-{x}"

        df[dummy_name] = dummies[x]

    df.drop(name, axis=1, inplace=True)
```

```python
encode_text_dummy(df,'http.request.method')

encode_text_dummy(df,'http.referer')

encode_text_dummy(df,"http.request.version")

encode_text_dummy(df,"dns.qry.name.len")

encode_text_dummy(df,"mqtt.conack.flags")

encode_text_dummy(df,"mqtt.protoname")
```

```python
df.info()
```

```python
y_column = df['Attack_label']
df.drop('Attack_label',axis=1,inplace=True)
df.drop('Attack_type',axis=1,inplace=True)
df.drop('mqtt.topic',axis=1,inplace=True)
dummies = pd.get_dummies(y_column) 
y = dummies.values
```

```python
# Standardisation des données
from sklearn.preprocessing import StandardScaler
X2 = StandardScaler().fit_transform(X)
```

```python
#Separation en train et test set
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X2, y,
                                                test_size = 0.4)
```

```python
xtrain.shape
```

```python
# creation d'un mlp a 5 couche
import tensorflow as tf
def mlp(input_dim):

    model = tf.keras.models.Sequential()

    # la premiere couche dense
    model.add(tf.keras.layers.Dense(8, input_dim=input_dim))

    # Quatre couches cachées
    model.add(tf.keras.layers.Dense(16, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(16, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(16, activation='sigmoid'))
    model.add(tf.keras.layers.Dense(32, activation='sigmoid'))

    # la couche de sortie
    model.add(tf.keras.layers.Dense(2, activation='sigmoid'))
    
    return model
```

```python
model_mlp = mlp(input_dim=xtrain.shape[1])

# Compile the model with mean squared error (for regression)
model_mlp.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
model_mlp.summary()
```

```python
from keras.utils.vis_utils import plot_model

plot_model(model_mlp, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model_mlp.fit(xtrain, ytrain, validation_data=(xtest, ytest), epochs=5, batch_size=128)
```

```python
from sklearn import metrics
y_pred_mlp = model_mlp.predict(X2)
y_pred_mlp = np.argmax(y_pred_mlp,axis=1) 
y_true_mlp = np.argmax(y,axis=1)

print("Accuracy: {}" .format(metrics.accuracy_score(y_true_mlp, y_pred_mlp)))
print("Precision: {}" .format(metrics.precision_score(y_true_mlp, y_pred_mlp, average='macro')))
print("Recall: {}" .format(metrics.recall_score(y_true_mlp, y_pred_mlp, average='macro')))
print("F1: {}" .format(metrics.f1_score(y_true_mlp, y_pred_mlp, average='macro')))
```

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_confusion_matrix
def plot_confusing_matrix (y_compare,pred,n_categories):

  cm = metrics.confusion_matrix(y_compare, pred)
  plot_confusion_matrix(conf_mat=cm,figsize=(13,13),show_normed=True)
  plt.title('Matrice de confusion')
  plt.ylabel('Cible')
  plt.xlabel('Prédite')
  plt.show()
```

```python
# Matrice de confusion
plot_confusing_matrix(y_true_mlp,y_pred_mlp,2)
```

proposition d'un model CNN-LSTM


```python
# Model CNN
from tensorflow import keras
from sklearn.model_selection import train_test_split
import numpy as np
from keras.utils import np_utils
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Conv2D, Flatten, MaxPooling2D,BatchNormalization
from keras.layers import LSTM,TimeDistributed,Conv1D,MaxPooling1D,Bidirectional
from keras.layers import Dense, Dropout, Flatten
from keras import models
from keras import layers
from keras.models import Model

CNN_LSTM = Sequential()
CNN_LSTM.add(Conv2D(32, kernel_size=(3, 3),activation='sigmoid',padding='same',input_shape=(1,92,1)))
CNN_LSTM.add(BatchNormalization())
CNN_LSTM.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
CNN_LSTM.add(Conv2D(16, (3, 3), activation='sigmoid',padding='same'))
CNN_LSTM.add(BatchNormalization())
CNN_LSTM.add(TimeDistributed(Flatten()))
CNN_LSTM.add(Bidirectional(LSTM(16,return_sequences=True,dropout=0.5,recurrent_dropout=0.5)))
CNN_LSTM.add(Bidirectional(LSTM(16,return_sequences=True,dropout=0.2)))
CNN_LSTM.add(Flatten())
CNN_LSTM.add(Dense(2, activation='sigmoid'))
CNN_LSTM.compile(loss='binary_crossentropy',optimizer='Adamax',metrics=['accuracy'])
CNN_LSTM.summary()
```

```python
from keras.utils.vis_utils import plot_model

plot_model(CNN_LSTM, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
history_cnn_lstm = CNN_LSTM.fit(X_train_cnn,ytrain,validation_data=(X_test_cnn,ytest),verbose=1,batch_size=500,epochs=2)
```

```python
from sklearn import metrics
y_pred_cnn_lstm = CNN_LSTM.predict(X_test_cnn)
y_pred_cnn_lstm = np.argmax(y_pred_cnn_lstm,axis=1) 
y_true = np.argmax(ytest,axis=1)

print("Accuracy: {}" .format(metrics.accuracy_score(y_true, y_pred_cnn_lstm)))
print("Precision: {}" .format(metrics.precision_score(y_true, y_pred_cnn_lstm, average='macro')))
print("Recall: {}" .format(metrics.recall_score(y_true, y_pred_cnn_lstm, average='macro')))
print("F1: {}" .format(metrics.f1_score(y_true, y_pred_cnn_lstm, average='macro')))
```

```python
from mlxtend.plotting import plot_confusion_matrix
def plot_confusing_matrix (y_compare,pred,n_categories):

  cm = metrics.confusion_matrix(y_true, y_pred_cnn)
  plot_confusion_matrix(conf_mat=cm,figsize=(13,13),show_normed=True)
  plt.title('Matrice de confusion')
  plt.ylabel('Cible')
  plt.xlabel('Prédite')
  plt.show()
```

```python
plot_confusing_matrix(y_true,y_pred_cnn_lstm,2)
```
